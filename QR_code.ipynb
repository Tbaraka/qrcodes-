```python
import qrcode
```

```python
url=input("Enter the URL").strip()
```

```python
file_path="my_qrcode.png" # You can change this to any desired filename or path, e.g., "output/my_qrcode.png"
qr=qrcode.QRCode()
qr.add_data(url)
img=qr.make_image()
img.save(file_path)
```
